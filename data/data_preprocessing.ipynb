```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#See: https://osf.io/nbuxg/ for the file, but feel free to just check out the preprocessed data
raw_data = "MasterAffectWithEventPresence.2019.csv"
raw_data_tab = pd.read_csv(raw_data)
```

```python
raw_data_tab = raw_data_tab.rename(columns = {raw_data_tab.columns[4]:"Affect_Rating_o", raw_data_tab.columns[12]:"Affect_Rating_1", 
                                       raw_data_tab.columns[25]:"Affect_Rating_2", raw_data_tab.columns[39]:"Affect_Rating_3"})
```

```python
newdata = {"Story_ID": [], "Emotion_Category": [], "Rating_o": [],
       "Rating_1": [], "Rating_2": [],
       "Rating_3": []}
```

```python
for i in raw_data_tab.index:
        if np.char.isnumeric(str(raw_data_tab.loc[i,"Number"])):
            newdata["Story_ID"].append(raw_data_tab.loc[i,"Original StoryID"])
            newdata["Emotion_Category"].append(raw_data_tab.loc[i,"Affect"].lower())
            newdata["Rating_o"].append(raw_data_tab.loc[i, "Affect_Rating_o"])
            newdata["Rating_1"].append(raw_data_tab.loc[i, "Affect_Rating_1"])
            newdata["Rating_2"].append(raw_data_tab.loc[i, "Affect_Rating_2"])
            newdata["Rating_3"].append(raw_data_tab.loc[i, "Affect_Rating_3"])
```

```python
newdata_tab = pd.DataFrame(newdata)
```
